```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from glob import glob
```

```python
path = '/home/mehul/Documents/data-analysis-data/data-analysis-data/weather_data_input/weather_data_set_*'
files = glob(path)
files
```

```python
dfs = [pd.read_csv(file, header=None, names=['stationno', 'year', 'datemonth', 'hourminute', 'temperature'], sep='|') \
 for file in files]
```

```python
dfs
```

```python
d1 = pd.concat(dfs, ignore_index=True)
d1
```

# maximum temperature recorded on yearly basis

```python
d2 = d1.ffill()
d2
```

```python
d3 = d2[['year', 'temperature']]
d3
```

```python
d4 = d3.groupby('year')
d4
```

```python
d5 = d4.max()
d5
```

```python
d5.plot(kind='bar')
plt.show()
```
