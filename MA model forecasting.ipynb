```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '/home/mehul/Documents/training/python/data/daily-min-temperatures.csv'
```

```python
df = pd.read_csv(file_path, parse_dates=['Date'], index_col=0)
df
```

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plot_acf(df)
plt.show()
```

```python
# q = 1
```

```python
x = df.values.ravel()
x
```

```python
partition = round(0.70 * len(x))
train, test = x[:partition], x[partition:]
```

```python
history = [x for x in train]
predictions = []
```

```python
from statsmodels.tsa.arima_model import ARMA
```

```python
for i in range(len(test)):
    model = ARMA(history, order=(1, 1)) # p = 0 (MA Model)
    model_fit = model.fit()
    prediction_list = model_fit.forecast()
    predictions.append(prediction_list[0][0])
    history.append(test[i])
    print('Predicted {0}'.format(prediction_list[0][0]))
    print('Expected {0}'.format(test[i]))
```

```python
plt.figure(figsize=(16, 5))
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from numpy import sqrt
```

```python
sqrt(mean_squared_error(test, predictions))
```
