```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
users = '/home/mehul/Documents/data-analysis-data/data-analysis-data/movielens/users.dat'
col_names = ['user_id', 'gender', 'age', 'occupation', 'zipcode']
users_df = pd.read_csv(users, header=None, names=col_names, sep='::')
users_df
```

```python
movies = '/home/mehul/Documents/data-analysis-data/data-analysis-data/movielens/movies.dat'
col_names = ['movie_id', 'title', 'genres']
movies_df = pd.read_csv(movies, header=None, names=col_names, sep='::')
movies_df
```

```python
ratings = '/home/mehul/Documents/data-analysis-data/data-analysis-data/movielens/ratings.dat'
col_names = ['user_id', 'movie_id', 'ratings', 'timestamp']
ratings_df = pd.read_csv(ratings, header=None, names=col_names, sep='::')
ratings_df
```

```python
d1 = (ratings_df.merge(movies_df)).merge(users_df)
d1
```

# popular films. A film where there is minimum sample size (number of people who voted) of 500

```python
s1 = d1['title']
s1
```

```python
s2 = s1.value_counts()
s3 = s2[s2 >= 500]
popular_films = s3.index
popular_films
```

```python
d2 = d1[['title', 'gender', 'ratings']]
d2
```

```python
d3 = d2.pivot_table('ratings', aggfunc='mean', index='title', columns=['gender'])
d4 = d3.loc[popular_films]
d5 = d4.sort_values(by='F', ascending=False)
d6 = d5.head(15)
d6
```

```python
d6.plot(kind='barh')
plt.show()
```

```python
d5['diff'] = d5['F'] - d5['M']
d5['diff'] = d5['diff'].abs()
d7 = d5.sort_values(by='diff', ascending=False)
d7
```
