```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
path = '/home/mehul/Documents/data-analysis-data/data-analysis-data/facebook_logs/facebook_login_logout_logs'
data = pd.read_csv(path, header=None, names=['username', 'date', 'time', 'action'])
data
```

# distinct set of users along with the frequency of the times that they have logged out that month

```python
d1 = data[['username', 'action']]
d1
```

```python
d2 = d1[d1['action'] == 'out']
d2
```

```python
s1 = d2['username']
s1
```

```python
s2 = s1.value_counts()
s2
```

```python
s2.plot(kind='bar')
plt.show()
```

# Alternate way to find the above

```python
d2
```

```python
# grouping and aggregation
d2['action'].count() # this is just aggregation

d3 = d2.groupby('username')
print(d3)
```

```python
d4 = d3.count() # this is aggregation on grouped data
d4
```

```python
d4.plot(kind='bar')
plt.show()
```
