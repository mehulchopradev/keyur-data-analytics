```python
from pandas import DataFrame
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import FactorAnalysis
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
```

```python
df = pd.read_csv(url, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'label'])
df
```

```python
x = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values
x
```

```python
scaler = StandardScaler()
std_normal_x = scaler.fit_transform(x)
std_normal_x
```

```python
print(std_normal_x.mean())
print(std_normal_x.std())
```

```python
analysis = FactorAnalysis()
analysis.fit(std_normal_x)
```

```python
analysis.components_
```

```python
d2 = DataFrame(analysis.components_, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
d2
```
