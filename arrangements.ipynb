```python
from itertools import permutations, combinations
```

```python
l = ['A', 'B', 'C', 'D', 'E', 'F']
ar = list(permutations(l, 6))
print(len(ar))
print(ar)
```

```python
l = ['A', 'B', 'C', 'D', 'E', 'F']
ar = list(permutations(l, 3))
print(len(ar))
print(ar)
```

```python
l = ['A', 'B', 'C', 'D', 'E', 'F']
ar = list(combinations(l, 6))
print(len(ar))
print(ar)
```

```python
l = ['A', 'B', 'C', 'D', 'E', 'F']
ar = list(combinations(l, 3))
print(len(ar))
print(ar)
```
