```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from json import loads
```

```python
path = '/home/mehul/Documents/data-analysis-data/data-analysis-data/url-shortner/usagov_bitly_data2012-03-16-1331923249.txt'
```

```python
fobj = open(path)
lines = []
for line in fobj:
    lines.append(loads(line))
lines
```

```python
d1 = DataFrame(lines)
d1
```

```python
s1 = d1['tz'].fillna('Missing')
s1[s1 == ''] = 'Missing'
s1
```

```python
d1['tz'] = s1
d1
```

```python
s2 = d1['tz']
s3 = s2.value_counts()
top_10 = s3.head(10)
top_10
```

```python
top_10.plot(kind='barh')
plt.show()
```

```python
d1['os'] = np.where(d1.a.str.contains('Windows'), 'Windows', 'Non-Windows')
d2 = d1[['os', 'tz']]
d2['t'] = 1
d2
```

```python
d3 = d2.pivot_table('t', aggfunc='sum', index='tz', columns=['os'])
d4 = d3.fillna(0)
d5 = d4.loc[top_10.index]
d5
```

```python
d5.plot(kind='bar')
plt.tight_layout()
plt.savefig('/home/mehul/Documents/url-shortner-analysis.jpg', format='jpg')
```
